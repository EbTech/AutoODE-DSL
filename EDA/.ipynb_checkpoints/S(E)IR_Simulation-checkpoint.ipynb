# SIR
\begin{align*}
&\frac{dS}{dt} = -\frac{\beta SI}{N} \\
&\frac{dI}{dt} = \frac{\beta SI}{N} - \gamma I \\
&\frac{dR}{dt} = \gamma I \\
& N = I + R + S
\end{align*}

```python
import torch 
import numpy as np
from scipy.integrate import odeint
# LSODA, written jointly with L. R. Petzold, solves systems dy/dt = f
# with a dense or banded Jacobian when the problem is stiff, but it
# automatically selects between nonstiff (Adams) and stiff (BDF) 
# methods.  It uses the nonstiff method initially, and dynamically 
# monitors data in order to decide which method to use.
import matplotlib.pyplot as plt
```

```python
def SIR(N, I0, R0, beta, gamma, num_steps):
    # Total population, N.
    N = N
    # Initial numbers, I0, R0, S0.
    I0, R0 = I0, R0
    S0 = N - I0 - R0

    # Contact rate, beta, and mean recovery rate, gamma.
    beta, gamma = beta, gamma#0.2, 1./10 

    # A grid of time points (half day)
    t = np.linspace(0, 60, num_steps + 1)

    # The SIR model differential equations.
    def deriv(y, t, N, beta, gamma):
        S, I, R = y
        dSdt = -beta * S * I / N
        dIdt = beta * S * I / N - gamma * I
        dRdt = gamma * I
        return dSdt, dIdt, dRdt

    # Initial conditions vector
    y0 = S0, I0, R0
    # Integrate the SIR equations over the time grid t.
    # if the ode is stiff, odeint use stiff(BDF) method. if the ode is non stiff, odeint use nonstiff adam method.
    ret = odeint(deriv, y0, t, args=(N, beta, gamma))
    S, I, R = ret.T
    return S, I, R, t
```

```python
S, I, R, t = SIR(N = 1000, I0 = 10, R0 = 0, beta = 0.9, gamma = 0.1, num_steps = 300)
N = 100000
```

```python
# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, S/N, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I/N, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R/N, 'g', alpha=0.5, lw=2, label='Recovered')
ax.set_xlabel('Time', size = 15)
ax.set_ylabel('Proportion', size = 15)
#ax.set_ylim(-0.1,1.2)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')

legend = ax.legend(fontsize = 15)
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
#plt.title("SIR:  " + r"$N=1000, I_0 = 1, \beta = 0.9, \gamma = 0.1$", fontsize = 20)
#plt.savefig("SIR.png", dpi = 400 , bbox_inches = "tight")    
plt.show()
```

# SEIR
\begin{align*}
&\frac{dS}{dt} = -\frac{\beta SI}{N} \\
&\frac{dE}{dt} = \frac{\beta SI}{N} - \sigma E \\
&\frac{dI}{dt} = \sigma E - \gamma I \\
&\frac{dR}{dt} = \gamma I \\
& N = I + R + S + E
\end{align*}

```python
def SEIR(N, I0, R0, E0, beta, gamma, sigma, num_steps):
    # Total population, N.
    N = N
    # Initial numbers, I0, R0, S0.
    I0, R0, E0 = I0, R0, E0 
    S0 = N - I0 - R0 - E0

    # Contact rate, beta, and mean recovery rate, gamma.
    beta, gamma, sigma = beta, gamma, sigma

    # A grid of time points 
    t = np.linspace(0, num_steps, num_steps+1)

    # The SIR model differential equations.
    def deriv(y, t, N, beta, gamma, sigma):
        S, E, I, R = y
        dSdt = -beta * S * I / N
        dEdt = beta * S * I / N - sigma * E
        dIdt = sigma * E - gamma * I
        dRdt = gamma * I
        return dSdt, dEdt, dIdt, dRdt

    # Initial conditions vector
    y0 = S0, E0, I0, R0
    # Integrate the SIR equations over the time grid t.
    # if the ode is stiff, odeint use stiff(BDF) method. if the ode is non stiff, odeint use nonstiff adam method.
    ret = odeint(deriv, y0, t, args=(N, beta, gamma, sigma))
    S, E, I, R = ret.T
    return S, E, I, R, t
```

```python
S, E, I, R, t = SEIR(N = 1000, I0 = 10, R0 = 0, E0 = 0, beta = 0.3, sigma = 0.4, gamma = 0.2,  num_steps = 30)
```

```python
N = 1000
# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, S/N, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, E/N, 'purple', alpha=0.5, lw=2, label='Exposed')
ax.plot(t, I/N, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R/N, 'g', alpha=0.5, lw=2, label='Recovered')
ax.set_xlabel('Time /days', size = 15)
ax.set_ylabel('Number (1000s)', size = 15)
#ax.set_ylim(-0.1,1.2)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend(fontsize = 15)
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.title("SEIR:  " + r"$N=1000, I_0 = 1, \beta = 0.9, \gamma = 0.1, \sigma = 0.2$", fontsize = 20)
#plt.savefig("SEIR.png", dpi = 400 , bbox_inches = "tight")    
plt.show()
```
