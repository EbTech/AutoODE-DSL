## SuEIR Piecewise

```python
from scipy.integrate import odeint
def SuEIR_Simu(N, E0, I0, R0, D0, beta, sigma, mu, gamma, step_size, t, num_steps = 10, S0 = None):
    u0 = (1-mu)*sigma*E0
    if S0 is None:
        S0 = N - I0 - R0 - E0 - u0

    # Contact rate, beta, and mean recovery rate, gamma.
    # A grid of time points (half day)
    # The SIR model differential equations.
    def deriv(y, t, N, beta, sigma, gamma, mu):
        S, E, I, R = y
        dSdt = -beta * S * (I + E) / N
        dEdt = beta * S * (I + E) / N - sigma * E
        dIdt = mu * sigma * E - gamma * I
        dRdt = gamma * I
        return dSdt, dEdt, dIdt, dRdt

    # Initial conditions vector
    y0 = S0, E0, I0, R0
    # Integrate the SIR equations over the time grid t.
    # if the ode is stiff, odeint use stiff(BDF) method. if the ode is non stiff, odeint use nonstiff adam method.
    ret = odeint(deriv, y0, t, args=(N, beta, sigma, gamma, mu))
    S, E, I, R = ret.T
    u = (1-mu)*sigma*E
    D = 0.2*R
    return S, u, E, I, R, D, t, N
```

```python
S1, u1, E1, I1, R1, D1, t1, N1 = SuEIR_Simu(N = 1000,  E0 = 10, I0 = 1, R0 = 0, D0 = 0, beta = 0.5, 
                                            sigma = 0.2, mu = 0.5, gamma = 0.1, step_size = 0.5, 
                                            t = np.arange(0, 60, 0.5)[:10], num_steps = 10)

S2, u2, E2, I2, R2, D2, t2, N2 = SuEIR_Simu(N = 1000, E0 = E1[-1], I0 = I1[-1], R0 = R1[-1], D0 = D1[-1], beta = 0.9, 
                                            sigma = 0.2, mu = 0.5, gamma = 0.1, step_size = 0.5, 
                                            t = np.arange(0, 60, 0.5)[9:20], num_steps = 10, S0 = S1[-1])

S3, u3, E3, I3, R3, D3, t3, N3 = SuEIR_Simu(N = 1000, E0 = E2[-1], I0 = I2[-1], R0 = R2[-1], D0 = D2[-1], beta = 0.7, 
                                            sigma = 0.2, mu = 0.5, gamma = 0.1, step_size = 0.5, 
                                            t = np.arange(0, 60, 0.5)[19:60], num_steps = 40, S0 = S2[-1])

S = np.concatenate([S1, S2[1:], S3[1:]])
u = np.concatenate([u1, u2[1:], u3[1:]])
E = np.concatenate([E1, E2[1:], E3[1:]])
I = np.concatenate([I1, I2[1:], I3[1:]])
R = np.concatenate([R1, R2[1:], R3[1:]])
D = np.concatenate([D1, D2[1:], D3[1:]])
t = np.arange(0, 60, 0.5)[:len(S)]

sample = torch.cat([torch.from_numpy(S).reshape(-1,1), 
                    torch.from_numpy(u).reshape(-1,1), 
                    torch.from_numpy(E).reshape(-1,1), 
                    torch.from_numpy(I).reshape(-1,1),
                    torch.from_numpy(R).reshape(-1,1),
                    torch.from_numpy(D).reshape(-1,1)], dim = 1).float()
torch.save(sample, "SuEIRD_piecewise_testcase.pt")
```

```python
class SuEIRD_Simu(nn.Module):
    def __init__(self, num_regions, solver = "Euler", symmetric = False, low_rank = None):
        super(SuEIRD_Simu, self).__init__()
        torch.manual_seed(1521) 
        self.init_E = torch.rand(50)/2 
        self.init_I = torch.rand(50)/10
        self.init_R = torch.rand(50)/20
        self.init_S = torch.ones(50).to(device) - self.init_I - self.init_R - self.init_E
        self.init_D = self.init_R*0.5
        
           #torch.ones(num_regions).to(device) - self.init_I - self.init_R - self.init_E

        if num_regions == 1:
            self.A = torch.ones(1, 1).to(device)
        else:
            if low_rank:
                if symmetric: 
                    self.B = torch.rand(num_regions, low_rank).to(device)
                    self.A = torch.mm(self.B, self.B.T)
                else:
                    self.B = torch.rand(num_regions, low_rank).to(device)
                    self.C = torch.rand(low_rank, num_regions).to(device)
                    self.A = torch.mm(self.B, self.C)
            else:
                self.A = torch.rand(num_regions, num_regions).to(device)
         
        
#         mask = torch.diag(torch.ones_like(self.A))
        self.A.data.fill_diagonal_(1)
#         torch.diagonal(self.A, 0) = 1.0
        
        self.beta = torch.rand(num_regions).to(device)
        self.gamma = torch.rand(num_regions).to(device)/2
        self.sigma = torch.rand(num_regions).to(device)/2
        self.mu = torch.rand(num_regions).to(device)/2
        self.step = torch.tensor(0.5).float().to(device)
        self.a = torch.rand(num_regions).to(device)/10
        self.b = torch.rand(num_regions).to(device)/2
        self.solver = solver
        
    def Euler(self, num_steps):
        S_pred = [self.init_S]
        E_pred = [self.init_E]
        I_pred = [self.init_I]
        R_pred = [self.init_R]
        D_pred = [self.init_D]
        for n in range(num_steps - 1):
            S_pred.append(S_pred[n] - self.beta * (torch.mm(self.A, ((I_pred[n] + E_pred[n]) * S_pred[n]).reshape(-1,1)).squeeze(1)) * self.step)
            E_pred.append(E_pred[n] + (self.beta * S_pred[n] * (I_pred[n]+ E_pred[n]) - self.sigma * E_pred[n]) * self.step)
            I_pred.append(I_pred[n] + (self.mu * self.sigma * E_pred[n] - self.gamma*I_pred[n]) * self.step)
            R_pred.append(R_pred[n] + self.gamma * I_pred[n] * self.step)
        for n in range(num_steps - 1):
            D_pred.append(D_pred[n] + self.a * torch.exp(- self.b * (n + 1) * self.step) * (R_pred[n+1] - R_pred[n]))
        y_pred = torch.cat([torch.stack(S_pred).transpose(0,1).unsqueeze(-1),
                            torch.stack(E_pred).transpose(0,1).unsqueeze(-1)*(1-self.mu)*self.sigma,
                            torch.stack(E_pred).transpose(0,1).unsqueeze(-1),
                            torch.stack(I_pred).transpose(0,1).unsqueeze(-1),
                            torch.stack(R_pred).transpose(0,1).unsqueeze(-1),
                            torch.stack(D_pred).transpose(0,1).unsqueeze(-1)], dim = -1)
        return y_pred
    
    def f_S(self, S_n, I_n, E_n): # TODO: Missing division by N =  I + R + S + E
        return -self.beta * ((I_n + E_n) * S_n) #-self.beta * (torch.mm(self.A, ((I_n + E_n) * S_n).reshape(-1,1)).squeeze(1))
    ##(torch.mm(self.A, .reshape(-1,1)).squeeze(1))
    
    def f_E(self, S_n, I_n, E_n): # TODO: Missing diving self.beta * S_n * (I_n + E_n) by N
        return self.beta * S_n * (I_n + E_n) - self.sigma * E_n
    
    def f_I(self, I_n, E_n):
        return self.mu * self.sigma * E_n - self.gamma*I_n
    
    def f_R(self, I_n):
        return self.gamma*I_n
    
    # dt is included in the ks
    def RK4_update(self, f_n, k1, k2, k3, k4):
        return f_n + 1/6 * (k1 + 2 * k2 + 2 * k3 + k4) * self.step
    
    def RK4(self, num_steps):
        S_pred = [self.init_S]
        E_pred = [self.init_E]
        I_pred = [self.init_I]
        R_pred = [self.init_R]
        D_pred = [self.init_D]
        for n in range(num_steps-1):
            # dt * f(t[n], y[n]) 
            k1_S = self.f_S(S_pred[n], I_pred[n], E_pred[n])
            k1_E = self.f_E(S_pred[n], I_pred[n], E_pred[n])
            k1_I = self.f_I(I_pred[n], E_pred[n])
            k1_R = self.f_R(I_pred[n])
            
            # dt * f(t[n] + dt/2, y[n] + k1/2)
            S_plus_k1_half = S_pred[n] + k1_S / 2 * self.step
            I_plus_k1_half = I_pred[n] + k1_I / 2 * self.step
            E_plus_k1_half = E_pred[n] + k1_E / 2 * self.step
            
            
            k2_S = self.f_S(S_plus_k1_half, I_plus_k1_half, E_plus_k1_half)
            k2_E = self.f_E(S_plus_k1_half, I_plus_k1_half, E_plus_k1_half)
            k2_I = self.f_I(I_plus_k1_half, E_plus_k1_half)
            k2_R = self.f_R(I_plus_k1_half)
            
            # dt * f(t[n] + dt/2, y[n] + k2/2)
            S_plus_k2_half = S_pred[n] + k2_S / 2 * self.step
            I_plus_k2_half = I_pred[n] + k2_I / 2 * self.step
            E_plus_k2_half = E_pred[n] + k2_E / 2 * self.step
            
            
            k3_S = self.f_S(S_plus_k2_half, I_plus_k2_half, E_plus_k2_half)
            k3_E = self.f_E(S_plus_k2_half, I_plus_k2_half, E_plus_k2_half)
            k3_I = self.f_I(I_plus_k2_half, E_plus_k2_half)
            k3_R = self.f_R(I_plus_k2_half)
            
            # dt * f(t[n] + dt, y[n] + k3) 
            S_plus_k3 = S_pred[n] + k3_S * self.step
            I_plus_k3 = I_pred[n] + k3_I * self.step
            E_plus_k3 = E_pred[n] + k3_E * self.step
            
            k4_S = self.f_S(S_plus_k3, I_plus_k3, E_plus_k3)           
            k4_E = self.f_E(S_plus_k3, I_plus_k3, E_plus_k3)
            k4_I = self.f_I(I_plus_k3, E_plus_k3) 
            k4_R = self.f_R(I_plus_k3)
            #print(I_pred[-1])
            S_pred.append(self.RK4_update(S_pred[n], k1_S, k2_S, k3_S, k4_S))
            E_pred.append(self.RK4_update(E_pred[n], k1_E, k2_E, k3_E, k4_E))
            I_pred.append(self.RK4_update(I_pred[n], k1_I, k2_I, k3_I, k4_I))
            R_pred.append(self.RK4_update(R_pred[n], k1_R, k2_R, k3_R, k4_R))
            
        for n in range(num_steps - 1):
            D_pred.append(D_pred[n] + (self.a * (n * self.step) + self.b) * (R_pred[n+1] - R_pred[n]))#self.a * torch.exp(-self.b * (n + 1) * self.step) * (R_pred[n+1] - R_pred[n]))

        y_pred = torch.cat([torch.stack(S_pred).transpose(0,1).unsqueeze(-1),
                            (torch.stack(E_pred)*(1-self.mu.unsqueeze(0))*self.sigma.unsqueeze(0)).transpose(0,1).unsqueeze(-1),
                            torch.stack(E_pred).transpose(0,1).unsqueeze(-1),
                            torch.stack(I_pred).transpose(0,1).unsqueeze(-1),
                            torch.stack(R_pred).transpose(0,1).unsqueeze(-1),
                            torch.stack(D_pred).transpose(0,1).unsqueeze(-1)], dim = -1)
        return y_pred
            
            
    def forward(self, num_steps):
        if self.solver == "Euler":
            return self.Euler(num_steps)
        elif self.solver == "RK4":
            return self.RK4(num_steps)
        else:
            print("Error")        
```
