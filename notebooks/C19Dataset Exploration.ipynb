# `C19Dataset` Exploration

Here we will make some visualizations of the data in our dataset, from the COVID-19 daily US reports.

```python
from ode_nn import C19Dataset
import matplotlib.pyplot as plt

plt.rc("font", family="serif", size=15)
%matplotlib inline
```

```python
ds = C19Dataset()
ds.df.head()
```

```python
fig, axes = plt.subplots(ncols=3, sharex=True, figsize=(12, 3))
for i, col in enumerate(["Confirmed", "Recovered", "Deaths"]):
    axes[i].plot(ds.df[col])
```

```python
df = ds.df.diff(axis=0)
fig, axes = plt.subplots(ncols=3, sharex=True, figsize=(12, 3))
for i, col in enumerate(["Confirmed", "Recovered", "Deaths"]):
    axes[i].plot(df[col])
```
